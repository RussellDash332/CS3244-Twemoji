```python
import pandas as pd
```

```python
train = pd.read_csv(r"https://raw.githubusercontent.com/RussellDash332/CS3244-Twemoji/main/Datasets/train_text_emoji_clean.csv")
test = pd.read_csv(r"https://raw.githubusercontent.com/RussellDash332/CS3244-Twemoji/main/Datasets/test_text_emoji_clean.csv")
valid = pd.read_csv(r"https://raw.githubusercontent.com/RussellDash332/CS3244-Twemoji/main/Datasets/valid_text_emoji_clean.csv")
```

```python
def to_columns(labels):
    labels = labels[1:-1].split(", ")
    labels = [int(label) for label in labels]
    if i in labels:
        return 1
    else:
        return 0
```

```python
for i in range(12):
    train[i] = train['labels'].apply(to_columns)
    test[i] = test['labels'].apply(to_columns)
    valid[i] = valid['labels'].apply(to_columns)
```

```python
train = train.drop("labels", axis = 1)
test = test.drop("labels", axis = 1)
valid = valid.drop("labels", axis = 1)
```

```python
train["labels"] = list(zip(*(train[i].tolist() for i in range(12))))
train["text"] = train["tweets"]
```

```python
test["labels"] = list(zip(*(test[i].tolist() for i in range(12))))
test["text"] = test["tweets"]
```

```python
valid["labels"] = list(zip(*(valid[i].tolist() for i in range(12))))
valid["text"] = valid["tweets"]
```

```python
from simpletransformers.classification import MultiLabelClassificationModel


model = MultiLabelClassificationModel('roberta', 'roberta-base', num_labels=12, use_cuda = False, args={'train_batch_size':2, 'gradient_accumulation_steps':16, 'learning_rate': 3e-5, 'num_train_epochs': 3, 'max_seq_length': 512})
#model_type may be one of ['bert', 'xlnet', 'xlm', 'roberta', 'distilbert'].
```

```python
model.train_model(train)
```
